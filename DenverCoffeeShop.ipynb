# A new coffee shop wants to open stores in Denver and believes that the ideal location is near affluent households and that their store appeals most to those aged 20-35.

## We want to find neighborhoods with high populations of people aged 20-35 near affluent neighborhoods.

### Don't have to worry about competition from Starbucks since the coffee shop doesn't particularly compete with them.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
coffee_df = pd.read_csv("/content/sample_data/denverCoffeeShop.csv")
```

```python
display(coffee_df)
```

```python
coffee_df[coffee_df['NUM_HHLD_100K+'] > 0].sort_values(by=['AGE_18_TO_34'], ascending=False)
```

# The chart above displays neighborhoods in Denver with the greatest target population for the Coffee Shop in descending order.

## The top three neighborhoods are **Capitol Hill** with a target population of 8,274, **Gateway - Green Valley Ranch** with a target population of 7,778, and **Montbello** with a target population of 7,742.

```python
coffee_df[coffee_df['NUM_HHLD_100K+'] > 0].sort_values(by=['NUM_HHLD_100K+'], ascending=False)
```

# The chart above displays neighborhoods in Denver with households whose income is above 100k in descending order.

## The top three neighborhoods are **Central Park** with 2,748 households above 100k in income, **Gateway - Green Valley Ranch** with 2,192 households above 100k in income, and **Hampden** with 2,160 households above 100k in income.

# Across both charts the **Gateway - Green Valley Ranch** neighborhood was a close second in target population and having large numbers of households with incomes above 100k.

## Therefore, the most optimal neighborhood for the coffee shop to open stores in should be **Gateway - Green Valley Ranch**.
